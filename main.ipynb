```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
from mpl_toolkits.mplot3d import Axes3D

# データセットをロード
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['label'] = iris.target

# ラベルごとのデータフレームを作成
df0 = data[data['label'] == 0]
df1 = data[data['label'] == 1]
df2 = data[data['label'] == 2]

xx = "sepal width (cm)"
yy = "sepal length (cm)"
zz = "petal length (cm)"

fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')

# データをプロット
ax.scatter(df0[xx], df0[yy], df0[zz], color="b")
ax.scatter(df1[xx], df1[yy], df1[zz], color="r")
ax.scatter(df2[xx], df2[yy], df2[zz], color="g")

# 軸ラベルを設定
ax.set_xlabel(xx)
ax.set_ylabel(yy)
ax.set_zlabel(zz)

# カメラの角度を設定
ax.view_init(0, 240)

plt.show()
```

```python
from sklearn.datasets import make_blobs
import pandas as pd

X, y = make_blobs(
random_state=3,
n_features=2,
centers=2,
cluster_std=1,
n_samples=300)

df = pd.DataFrame(X)
df["target"] = y
df.head()
```

```python
X, y = make_blobs(
random_state=3,
n_features=2,
centers=5,
cluster_std=1,
n_samples=300)

df = pd.DataFrame(X)
df["target"] = y
# 分類によって、別々のデータフレームに分ける
df0 = df[df["target"]==0]
df1 = df[df["target"]==1]
df2 = df[df["target"]==2]
df3 = df[df["target"]==3]
df4 = df[df["target"]==4]
# 分類0は青、1は緑、3はマゼンダ、4はシアンで、散布図を描画
plt.figure(figsize=(5, 5))
plt.scatter(df0[0], df0[1], color="b", alpha=0.5)
plt.scatter(df1[0], df1[1], color="r", alpha=0.5)
plt.scatter(df2[0], df2[1], color="g", alpha=0.5)
plt.scatter(df3[0], df3[1], color="m", alpha=0.5)
plt.scatter(df4[0], df4[1], color="c", alpha=0.5)
plt.grid()
plt.show()
```
